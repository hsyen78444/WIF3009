**Question 13: Non-Continuous Data Extraction**  
You have a sequential list of mixed data points, but you only need to extract specific elements that are not in a continuous range (specifically, elements at indices 1, 3, and 5).

* **A)** Without using external libraries like NumPy, write the single line of code demonstrated in the lecture to extract a new list containing only the floats `[1.24, 1.67, 1.98]`.

**Question 16: Complex Logic & Iteration**  
You are filtering a list of patient ages to find those eligible for a specific treatment. 
`ages = [18, 25, 30, 45, 60]`
The eligibility criteria dictate that a patient must be strictly greater than 20 AND less than or equal to 45.
* **A)** Write a `for` loop combined with an `if-else` statement that iterates through the `ages` list. For each age, it should print `'Eligible'` if the compound condition is met, and `'Not Eligible'` otherwise.

```python
ages = [18, 25, 30, 45, 60]
for age in ages:
    if age > 20 and age <= 45:
        print('Eligible')
    else:
        print('Not Eligible')
```
