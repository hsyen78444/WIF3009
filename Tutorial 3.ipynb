### Question 3

The exercise will involve data cleaning, manipulation, and analysis tasks, including the use of the pivot() and melt() functions in Pandas:  

1. Load the data from the CSV file.
2. Perform data cleaning tasks such as handling missing values and data types.
3. Create additional columns for TotalPrice and YearMonth from the Timestamp column.
4. Aggregate the data to calculate total sales for each store and region.
5. Use the pivot() function to create a summary table showing total sales for each ProductName across different regions.
6. Use the melt() function to reshape the original data into a long format. The id columns are 'StoreID', 'Region', 'ProductID'
7. Save the manipulated data to new CSV files.

````python
data = pd.read_csv('sales_transactions.csv')
data.info()
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data.info()
````

````python
#Create additional columns for TotalPrice and YearMonth from the Timestamp column.
data['TotalPrice'] = data['Quantity'] * data['Price']
data['YearMonth'] = data['Timestamp'].dt.to_period('M')
data
````

````python
#Aggregate the data to calculate total sales for each store and region.

total_sales = data.groupby(['StoreID', 'Region'])['TotalPrice'].sum()
total_sales
````

````python
#Use the pivot() function to create a summary table showing total sales for each ProductName across different regions.
pivot_table = data.pivot(index='ProductName', columns='Region', values='TotalPrice').fillna(0)
pivot_table
````

````python
#Use the melt() function to reshape the original data into a long format. The id columns are 'StoreID', 'Region', 'ProductID'

melted_data = data.melt(id_vars=['StoreID', 'Region', 'ProductID'], var_name='Variable', value_name='Value')
melted_data
# Save the manipulated data to new CSV files
total_sales.to_csv('total_sales.csv')
pivot_table.to_csv('pivot_table.csv')
melted_data.to_csv('melted_data.csv', index=False)
````

### Question 4

The Excel file contains two sheets:

Student Scores: Contains columns for StudentID, Name, MathScore, PhysicsScore, ChemistryScore, and BiologyScore. Each row represents a student's scores in different subjects.  
Student Details: Contains additional details about each student, including StudentID, Name, Gender, Age, and Grade.

Task Description:

1. Load the data from the Student Scores and Student Details sheets in the Excel file.
2. Perform data cleaning tasks such as handling missing values and data types.
3. Merge the scores data with student details using StudentID as the key.
4. Calculate the average score for each student and add it as a new column.
5. Use the groupby() function to calculate average scores for each subject.
6. Use the pivot_table() function to create a summary table showing average scores for each subject by gender.
7. Use the melt() function to reshape the merged data into a long format.
    * ID columns are = 'StudentID', 'Name', 'Gender', 'Age', 'Grade'.
    * Melted columns are = 'MathScore', 'PhysicsScore', 'ChemistryScore', 'BiologyScore'
9. Save the manipulated data to a new Excel file.

````python
!pip install openpyxl

scores_df = pd.read_excel('student_performance.xlsx', sheet_name='Student Scores')
details_df = pd.read_excel('student_performance.xlsx', sheet_name='Student Details')
````

````python
print("Student Scores:")
print(scores_df.head())

print("\nStudent Details:")
print(details_df.head())
````

````python
# Check missing values
print("\nMissing values in Student Scores:")
print(scores_df.isnull().sum())

print("\nMissing values in Student Details:")
print(details_df.isnull().sum())
````

````python
print(scores_df.info())
````

````python
print(details_df.info())
````

````python
#Merge the scores data with student details using StudentID as the key.
merged_df = pd.merge(scores_df, details_df, on='StudentID', how='inner')
print(merged_df.head())
````

````python
#Calculate the average score for each student and add it as a new column
merged_df['AvgScore'] = merged_df[['MathScore', 'PhysicsScore', 'ChemistryScore','BiologyScore']].mean(axis=1)
merged_df
````

````python
#Use the groupby() function to calculate average scores for each subject.
subject_avg = merged_df[['MathScore', 'PhysicsScore', 'ChemistryScore','BiologyScore']].mean()
subject_avg.columns = ['Subject', 'AverageScore']
subject_avg
````

````python
#Use the pivot_table() function to create a summary table showing average scores for each subject by gender.
pivot_summary = pd.pivot_table(
    merged_df,
    values=['MathScore', 'PhysicsScore', 'ChemistryScore','BiologyScore'],
    index='Gender',
    aggfunc='mean'
)

pivot_summary
````

````python
#Use melt() to reshape the merged data into long format, 
# * ID columns are = 'StudentID', 'Name', 'Gender', 'Age', 'Grade'. 
# Melted columns are = 'MathScore', 'PhysicsScore', 'ChemistryScore', 'BiologyScore'

melted_df = pd.melt(
    merged_df,
    id_vars=['StudentID', 'Name', 'Gender', 'Age', 'Grade'],
    value_vars=['MathScore', 'PhysicsScore', 'ChemistryScore', 'BiologyScore'],
    var_name='Subject',
    value_name='Score'
)

melted_df.head(101)
````

````python
with pd.ExcelWriter('student_performance_analysis.xlsx') as writer:
    merged_df.to_excel(writer, sheet_name='MergedData', index=False)
    subject_avg.to_excel(writer, sheet_name='SubjectAverages')
    pivot_summary.to_excel(writer, sheet_name='PivotSummary')
    melted_df.to_excel(writer, sheet_name='MeltedData', index=False)
    
````
