### Question 1

Based on the data below, do the following:  

```
# Sample data for df1
data1 = {
    'user_id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
}

# Sample data for df2
data2 = {
    'user_id': [2, 3, 4],
    'age': [25, 30, 35]
}
```
    
1. Import necessary libraries
2. Load the data into a dataframe
3. Concatenate the dataframes (vertically)
4. Concatenate the dataframes (horizontally)
5. Merge the dataframes on the column `user_id` using inner join.
6. Join the dataframes with `user_id` as the index. This time, do an outer join.

````python
import pandas as pd

# Sample data for df1
data1 = {
    'user_id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
}

# Sample data for df2
data2 = {
    'user_id': [2, 3, 4],
    'age': [25, 30, 35]
}

# Creating DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

df_concatenated_vertical = pd.concat([df1, df2], ignore_index=True, sort=False)
print("Vertically Concatenated DataFrame:")
print(df_concatenated_vertical)

df_concatenated_horizontal = pd.concat([df1, df2], axis=1)
print("Horizontally Concatenated DataFrame:")
print(df_concatenated_horizontal)

df_merged = pd.merge(df1, df2, on='user_id', how='inner')
print("Merged DataFrame (Inner Join):")
print(df_merged)

df_concatenated_vertical = pd.concat([df1, df2], ignore_index=True, sort=False)
print("Vertically Concatenated DataFrame:")
print(df_concatenated_vertical)
````

### Question 2

Based on the data below, do the following:  

```
grades = {
    'StudentID': ['S1', 'S2', 'S3'],
    'Math': [88, 75, 92],
    'Science': [84, 88, 93]
}

more_grades = {
    'StudentID': ['S2', 'S3', 'S4'],
    'Literature': [78, 85, 88],
    'Art': [82, 90, 85]
}

activities = {
    'StudentID': ['S1', 'S2', 'S4'],
    'Club': ['Robotics', 'Math Club', 'Art Club']
}
```
    
1. Import necessary libraries
2. Load the data into a dataframe
3. Append `grades` and `more_grades` row-wise.
4. Merge `grades` and `activities` on `StudentID` to see grades and club participation.
5. Use join() to merge `grades` and `activities` based on `StudentID` column as index.

````python
import pandas as pd

# Creating grades DataFrame
grades_df = pd.DataFrame({
    'StudentID': ['S1', 'S2', 'S3'],
    'Math': [88, 75, 92],
    'Science': [84, 88, 93]
})

# Creating more grades DataFrame
more_grades_df = pd.DataFrame({
    'StudentID': ['S2', 'S3', 'S4'],
    'Literature': [78, 85, 88],
    'Art': [82, 90, 85]
})

# Creating activities DataFrame
activities_df = pd.DataFrame({
    'StudentID': ['S1', 'S2', 'S4'],
    'Club': ['Robotics', 'Math Club', 'Art Club']
})

# Row-wise concatenation
combined_grades_df = pd.concat([grades_df, more_grades_df], ignore_index=True)
print(combined_grades_df)

# Merging on StudentID
grades_activities_df = pd.merge(grades_df, activities_df, on='StudentID', how='left')
print(grades_activities_df)

# Setting the index to StudentID for joining
grades_df_indexed = grades_df.set_index('StudentID')
activities_df_indexed = activities_df.set_index('StudentID')

# Joining on the index
joined_df = grades_df_indexed.join(activities_df_indexed, how='left')
print(joined_df)
````

Note: Make sure figsize of the image is (10,6)

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
months = np.array(["January", "February", "March", "April", "May", "June", 
                   "July", "August", "September", "October", "November", "December"])
aqi_city_a = np.random.randint(50, 150, size=12)
aqi_city_b = np.random.randint(60, 160, size=12)
aqi_city_c = np.random.randint(40, 140, size=12)
aqi_city_d = np.random.randint(70, 170, size=12)
````

````python
fig1 = plt.figure(figsize=(10, 6))
plt.plot(months, aqi_city_a, marker='o', linestyle='-', color='blue', label='City A')
plt.plot(months, aqi_city_b, marker='s', linestyle='-', color='red', label='City B')
plt.plot(months, aqi_city_c, marker='^', linestyle='-', color='green', label='City C')
plt.plot(months, aqi_city_d, marker='d', linestyle='-', color='purple', label='City D')
plt.title('Monthly AQI Comparison Among Four Cities')
plt.xlabel('Month')
plt.ylabel('AQI')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
fig2 = plt.figure(figsize=(10, 6))
plt.scatter(aqi_city_a, aqi_city_b, color='darkorange')
plt.title('AQI Comparison Between City A and City B')
plt.xlabel('AQI City A')
plt.ylabel('AQI City B')
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
fig3 = plt.figure(figsize=(10, 6))
plt.hist(aqi_city_a, bins=6, color='skyblue', alpha=0.7)
plt.title('AQI Distribution for City A')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
````

````python
average_aqi = [np.mean(aqi_city_a), np.mean(aqi_city_b), np.mean(aqi_city_c), np.mean(aqi_city_d)]
cities = ['City A', 'City B', 'City C', 'City D']

fig4 = plt.figure(figsize=(10, 6))
plt.bar(cities, average_aqi, color=['blue', 'red', 'green', 'purple'])
plt.title('Average AQI Comparison Among Four Cities')
plt.xlabel('City')
plt.ylabel('Average AQI')
plt.tight_layout()
plt.show()
````

The dataset will simulate geographical and environmental data for a set of locations. Each location has the following attributes:

Use the code below to generate the synthetic dataset:

# Generate a dataset

longitude = np.random.uniform(-180, 180, 100)

df = pd.DataFrame({

Based on the dataset, perform the following operations:  

1. Plot geographical points with their elevation and use temperature to color-code the points. Use `scatter_3d`.
2. A 3D line chart could be used to represent a path or journey. Simulate a path by sorting the data by elevation and plotting the result. Use `line_3d`.
3. 3D surface plots are useful for representing topographical surfaces or any kind of 3D surface data.  
   Create a simplified surface plot simulating a mountain range.  Use the `plotly.graph_objects`.  Reference can get from here: https://plotly.com/python/3d-surface-plots/#topographical-3d-surface-plot  
   For the purposes of this example, we'll generate a structured grid of points and compute a hypothetical elevation.
   Use the data below for the third task:

```
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
```

````python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate a dataset
np.random.seed(42) # Ensure reproducibility
longitude = np.random.uniform(-180, 180, 100)
latitude = np.random.uniform(-90, 90, 100)
elevation = np.random.uniform(0, 8848, 100) # Max height to Mount Everest
temperature = np.random.uniform(-50, 50, 100) # Temperature range

df = pd.DataFrame({
    'Longitude': longitude,
    'Latitude': latitude,
    'Elevation': elevation,
    'Temperature': temperature
})
````

````python
fig = px.scatter_3d(df, x='Longitude', y='Latitude', z='Elevation',
                    color='Temperature',
                    color_continuous_scale=px.colors.sequential.Viridis,
                    title='3D Scatter Plot of Geographical Points')
fig.show()
````

````python
df_sorted = df.sort_values(by='Elevation')

fig = px.line_3d(df_sorted, x='Longitude', y='Latitude', z='Elevation',
                 title='3D Line Chart Representing an Ascending Path')
fig.show()
````

````python
import plotly.graph_objects as go

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot Simulating a Mountain Range', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
````
