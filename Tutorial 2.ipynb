## Part 1: NumPy Basics & Element-wise Operations
**Scenario:** You have two lists containing the heights (in meters) and weights (in kg) of three patients. Regular Python lists do not support element-wise operations. You need to calculate the BMI for all patients at once, which requires NumPy.

**Question 1:** Import the NumPy package, convert the lists into NumPy arrays, and calculate the element-wise division of weights by heights to find the BMI.

```python
heights = [1.50, 1.75, 1.82]
weights = [55, 70, 85]
```

````python
import numpy as np
heights = np.array([1.50, 1.75, 1.82])
weights = np.array([55, 70, 85])
bmi = weights / heights**2
print(bmi)
````

Question 2: Using your np_weights array, extract the weights of patients who weigh more than 60 kg but less than 80 kg. Remember that NumPy arrays require specific boolean operators (& instead of and).

## Part 7: File Handling
Scenario: You need to maintain a log file of patient visits.


Question 7: 
1. Open a file named log.txt in write mode (`'w'`) to create it and write "Patient 1 visited.\n". Close the file.
2. Open the same file in append mode (`'a'`) and add "Patient 2 visited.". Close it.
3. Open it in read mode (`'r'`) and print the lines.

````python
with open("log.txt", "w") as f:
    f.write("Patient 1 visited.\n")
    f.close()
with open("log.txt", "a") as f:
    f.write("Patient 2 visited.")
    f.close()
with open("log.txt", "r") as f:
    print(f.read())
````

## Part 8: Classes

Scenario: To better organize patient data, you decide to use Object-Oriented Programming.

Question 8: Create a Patient class. It should have an `__init__()` function that assigns name and age properties. Include a `speak()` method that prints `"I am [name]"`. Instantiate an object, change its age (mutating), and then delete the age property.

````python
class Patient:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def speak(self):
        print("I am " + self.name)

patient1 = Patient("Alice", 30)
patient1.age = 51
del patient1.age
````
